```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
pidcount = purchase_data["Purchase ID"].value_counts()
pidcount = len(pidcount)
pidcount
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueitems = purchase_data["Item Name"].unique()
uniqueitemslist = list(uniqueitems)
uniqueitemslistcount = len(uniqueitemslist)
uniqueitemslistcount

AVGPrice = purchase_data["Price"].mean()
AVGPriceF = '${:,.2f}'.format(AVGPrice)
AVGPrice



TotalSales = purchase_data["Price"].sum()
TotalSalesF = '${:,.2f}'.format(TotalSales)

Summary = {"Number of Unique Items": uniqueitemslistcount,
            "Average Price":  AVGPriceF,
            "Total Sales": TotalSalesF}
Summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Gcount = purchase_data["Gender"].value_counts()
Gcount

MCount = purchase_data.loc[purchase_data["Gender"] == "Male", :].value_counts()
MCounts = len(MCount)
MCounts
MPer = MCounts/pidcount
MPerF = "{:.2%}".format(MPer)
MPerF

FCount = purchase_data.loc[purchase_data["Gender"] == "Female", :].value_counts()
FCounts = len(FCount)
FCounts
FPer = FCounts/pidcount
FPer
FPerF = "{:.2%}".format(FPer)
FPerF


NaCount = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :].value_counts()
NaCounts = len(NaCount)
NaCounts
NaPer = NaCounts/pidcount
NaPer
NaPerF = "{:.2%}".format(NaPer)
NaPerF

GenderDF = pd.DataFrame({"Amount of Male Player": [MCounts, MPerF],
                        "Amount of Females Players": [FCounts, FPerF],
                        "Amount of Other / Non-Disclosed": [NaCounts, NaPerF]})
GenderDF.index = ["Count", "Percentage"]
GenderDF

```

```python
Agebin = [0,9,14,19,24,29,34,39, 1000]
Agelabels = ["0-9", "10-14"," 15-19", "20 - 24", " 25-29", "30-34", "35-39", "40 and up"]

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], Agebin, labels=Agelabels, include_lowest=True)
# purchase_data

Age_group = purchase_data.groupby("Age Range")
# AgeGroupC = Age_group[["Age", "Price"]].count()
AgeGroupM = Age_group[["Age", "Price"]].mean()
AgeGroupM["Price"] = AgeGroupM["Price"].map("${:.2f}".format)
AgeGroupM["Age"] = AgeGroupM["Age"].round(2)

AgeGroupM.rename(columns = {"Age":"Average Age", "Price" : "Average Price"}, inplace = True)
AgeGroupM
# AgeGroupC
# pidcount
```
